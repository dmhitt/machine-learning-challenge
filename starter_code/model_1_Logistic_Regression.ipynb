```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
#selected_features = df.drop(['koi_disposition'], axis = 1)
X = df.drop(['koi_disposition'], axis = 1)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df['koi_disposition']
y
#df.koi_disposition.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
#y_train
```

```python
X_train
```

```python
y_train
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)
#y_minmax = MinMaxScaler().fit(y_train)

X_train_scaled = X_minmax.transform(X_train)
X_test_scaled = X_minmax.transform(X_test)
#y_train = y_minmax.transform(y_train)
#y_test = y_minmax.transform(y_test)
```

# Train the Model



```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='liblinear', random_state=0, multi_class ='ovr')
# 'multi_class' : ['auto', 'ovr', 'multinomial']
classifier
```

```python
classifier.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")

```

```python
#Dinnara
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()

# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50], 
              'solver' : ['newton-cg', 'liblinear'],
              'multi_class' : ['auto', 'ovr', 'multinomial'] }
    
grid = GridSearchCV(classifier, param_grid, verbose=3, return_train_score=True)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
#Dinnara
predictions = grid.predict(X_test)
```

```python
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["FALSE POSITIVE", "CONFIRMED", "CANDIDATE"]))
```

```python
results = pd.DataFrame(grid.cv_results_)
results
```
