```python
from rfpimp import *
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
df
```

```python
df.koi_disposition.value_counts()
```

```python
#df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.


X = df.drop(['koi_disposition', 'koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec'], axis = 1)
#X = df.drop(['koi_disposition'], axis = 1)
X.columns
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python

y = df['koi_disposition']
y = pd.get_dummies(y)
y = y.drop(['CANDIDATE', 'FALSE POSITIVE'], axis = 1)
y = np.ravel(y)
y
```

```python
# Create a random forest classifier
#rf = RandomForestClassifier(n_estimators=200)
rf = RandomForestClassifier()
from sklearn.model_selection import train_test_split
X_t, X_v, y_t, y_v = train_test_split(X, y)
rf = rf.fit(X_t, y_t)
print(rf.score(X_t, y_t), rf.score(X_v, y_v))
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
#it tells you that the 4th feature is the most important
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, X.columns), reverse=True)
```
